```python
#Profit Analysis by City:
```

```python
city = df.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()
```

```python
city = city.sort_values('Profit')
```

```python
city.head()
```

```python
# High profit analysis

city['Profit'].head(20).plot(kind='bar',figsize=(15,5),color = 'Red')
plt.title('Sales, Discount and Profit Analysis by City')
plt.show()
```

```python
# Low profit analysis

city['Profit'].tail(20).plot(kind='bar',figsize=(15,5),color = 'Yellow')
plt.title('Sales, Discount and Profit Analysis by City')
plt.show()
```

```python
# Sales, Profit and Discount Analysis by Region
```

```python
region_ = df.groupby(['Region'])[['Sales', 'Discount', 'Profit']]
```

```python
region_ = region_.mean()
```

```python
region_.head()
```

```python
region_.plot.pie(subplots=True, figsize=(12, 10), autopct='%1.1f%%', labels = region_.index)
plt.title('Region Wise Profit Analysis', fontsize = 10)
plt.legend()
plt.show()
```
